## Part 3: Subsets of similar opcodes

```octave
f = fopen('files.txt','r');
files = textscan(f, '%s');
files = files{1};
fclose(f);

m = length(files);
all = {};
for k=1:m
    fn=strcat('exmp/',files{k});
    f = fopen(fn, 'r');
    fileData(k) = textscan(f, '%s');
    fclose(f);
    all = vertcat(all, fileData{k});
end
```
